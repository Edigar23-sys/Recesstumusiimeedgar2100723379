<p style="color: blue;">1.Write a Pandas program to compare the elements of two Pandas series.  </p>

```python

import pandas as pd

# Creating two series
series1 = pd.Series([4, 65, 436, 3, 9])
series2 = pd.Series([7, 0, 3, 897, 9])

# Compare the elements of the two series
comparison_result = series1 == series2

# Print the comparison result
print(comparison_result)
```

<p style="color: blue;">2.Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
Sample Series: [2,4,6,8,14], [1,3,5,7,9]
  </p>

```python

#
import pandas as pd

# Create the two series
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

# Perform arithmetic operations on the series
addition = series1 + series2
subtraction = series1 - series2
multiplication = series1 * series2
division = series1 / series2

# Print the results
print("Addition:")
print(addition)
print("-------------------------")
print("\nSubtraction:")
print(subtraction)
print("-------------------------")
print("\nMultiplication:")
print(multiplication)
print("-------------------------")
print("\nDivision:")
print(division)
print("-------------------------")
```

Sample dictionary: dictionary1 = {‘Josh’: 24, ‘Sam’: 36, ‘Peace’: 19, ‘Charles’: 65, ‘Tom’: 44}

<p style="color: blue;">  4.Write a Pandas program to convert a given series to an array.
 Sample series: [‘Love’, 800, ‘Joy’, 789.9, ‘Peace’, True]

<p style="color: blue;"> 5.Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series.
(Use the ‘HomeTeamGoals’ column in the dataset provided)</p>

```python

import pandas as pd

df =pd.read_csv('AfricaCupofNationsMatches.csv')

# Get the most frequent value in the 'HomeTeamGoals' column
most_frequent_value = df['HomeTeamGoals'].value_counts().idxmax()

# Replace everything else as 'Other'
df.loc[df['HomeTeamGoals'] != most_frequent_value, 'HomeTeamGoals'] = 'Other'

# Print the resulting DataFrame
print(df['HomeTeamGoals'])
```
